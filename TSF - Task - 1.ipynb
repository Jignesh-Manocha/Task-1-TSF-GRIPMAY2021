```python
#Required Libraries for importing the data
import pandas as pd

#Required Libraries for data visualization
import seaborn as sns
from matplotlib import pyplot as plt

#Required Libraries for training and testing the data using Linear Regression Model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#Required Libraries for Model Evaluation
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

# Importing the Data

```python
#importing the data

df = pd.read_csv("http://bit.ly/w-data")
df.head(20) 
```

# Data Visualization

```python
#visualizing the relation between the features using pairplot

sns.pairplot(df)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(
    df['Hours'],
    df['Scores'],
    c='black'
)
plt.xlabel("Hours")
plt.ylabel("Scores)")
plt.show()
```

# Model Training

```python
X = df.drop('Scores', axis = 1)
Y = df.Scores

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
reg = LinearRegression().fit(X_train, y_train) 
```

# Model Testing

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

# Checking the Accuracy

```python
print('Training Accuracy: ', reg.score(X_train, y_train))
print('Testing Accuracy: ', reg.score(X_test, y_test))
```

```python
predictions = reg.predict(X)
plt.figure(figsize=(6, 4))
plt.scatter(
    df['Hours'],
    df['Scores'],
    c='black'
)
plt.plot(
    df['Hours'],
    predictions,
    c='blue',
    linewidth=2
)
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

# Making Prediction

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df
```

```python
print('If a student studies for 9.25 hrs/day, his predicted score would be', float(reg.predict([[9.25]]).round(3)))
```

# Model Evaluation

```python
mean_squ_error = mean_squared_error(y_test, y_pred)
mean_abs_error = mean_absolute_error(y_test, y_pred)
print("Mean Squred Error:",mean_squ_error)
print("Mean absolute Error:",mean_abs_error)
```
